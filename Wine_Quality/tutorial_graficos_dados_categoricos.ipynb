#Tipos de grficos

Como o objetivo desse dataset  a classificao dos vinhos, nosso primeiro grfico vai ser uma avaliao da coluna de qualidade. Essa avaliao pode ser por meio de um grfico dos seguintes tipos:

1) histograma

# Histograma



### Matplotlib

No seguinte link, podemos encontrar uma explicao mais detalhada da biblioteca matplotlib que foi a biblioteca utilizada para a criao dos seguintes grficos.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist

De maneira bem simples, podemos simplesmente utilizar o comando .hist() para produzirmos um grfico do tipo histograma.



```python
# the histogram of the data
'''
matplotlib.pyplot.hist(x, bins=None, range=None, density=False, weights=None, cumulative=False,
                        bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None,
                          log=False, color=None, label=None, stacked=False, *, data=None, **kwargs)
'''
plt.hist(wine_df['quality'])
plt.show()
```

No entanto, o grfico no possui nenhuma informao com relao aos eixos, no possui um ttulo e muitas das vezes a colorao do grfico pode estar em conflito com a colorao do restante do material.

O parmetro ***bins***  referente a quantidade de colunas que queremos apresentar, por exemplo caso bins=2 vamos ter apenas duas colunas para representar todos os dados. O problema que isso pode acarretar  mascarar as informaes onde uma mesma coluna pode ser relacionada a mais de uma classificao.

O parmetro ***facecolor*** permite fazer alteraes com relao a cor do grfico, e o parmetro ***alpha*** est relacionado a opacidade da cor do grfico sendo 1 a cor natural e 0 branco.

```python
'''
Link onde podemos encontrar outras cores para nosso grfico

https://matplotlib.org/stable/tutorials/colors/colors.html

'''
plt.hist(wine_df['quality'],bins=10,facecolor='g', alpha=0.3)
plt.show()
```

Para melhorar ainda mais o entendimento sobre o grfico, podemos inserir os ttulos nos eixos assim como um ttulo para o grfico.


```python
plt.hist(wine_df['quality'],10,facecolor='g', alpha=0.3)
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.title('Histogram of wine quality')
plt.show()
```

# Grfico tipo pizza ou torta

### agrupamento de dados

Para conseguir fazer o grfico de pizza vamos precisar fazer um agrupamento dos dados de interesse. Esse agrupamento dos dados podem ser feito das seguintes maneiras:

1) df['col'].value_counts().sort_values()

fazemos um filtro na coluna desejada, em seguinda fazemos uma contagem das quantidades de vezes que aparece cada valor e em seguida fazemos uma ordenao de valores.


2)df.groupby('col')['col'].count()

fazemos um agrupamento levando em conta a 'col', em seguida aplicamos um filtro para exibir apenas a 'col' e por fim fazemos a contagem desejada

```python
wine_df['quality'].value_counts().sort_values()
```

```python
#groupby the data by delivery type
data = wine_df.groupby("quality")['quality'].count()
data
```

Agora vamos criar duas listas contendo a quantidade de vezes da apario de cada classificao.

aparies: values
nomes das qualidades: label

```python
label=[3,4,5,6,7,8]
values=[10,53,681,638,199,18]
```

### Matplotlib

Fazendo o grfico introduzindo os valores percentuais no mesmo, podemos observar que o grfico gerado ficou muito pouludo.

Para a criao de grficos, precisamos sempre ter a grande mxima na mente, onde "menos  mais".

```python
plt.pie(values, labels=label,autopct = '%0.1f%%') #esse valor do autopct  para exibir as casa decimais do valor percentual
plt.show()
```

Removendo os valores de dentro grfico, e intoduzindo a legenda, ainda temos que fazer mais ajustes referentes ao posicionamento da legenda e o tamanho do grfico.

```python
explode = [0.3,0.02,0.02,0.02,0.02,0.02]
fig1, ax1 = plt.subplots()
ax1.pie(values, explode=explode, labels=label)
ax1.legend(title = "Wine quality:")

plt.show()
```

Aumentando o tamanho do grficos, melhorando a posio da legenda conseguimos produzir um grfico que melhor facilita sua compreenso.

```python
explode = (0.025, 0.025, 0.05, 0.025 ,0.025 ,0.025)
fig1, ax1 = plt.subplots(figsize=(10, 10))
ax1.pie(values, explode=explode, labels=label,
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.legend(title ="Wine Quality",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
```

# Grficos de barra

### Seaborn

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white", context="talk")
rs = np.random.RandomState(8)

# Set up the matplotlib figure
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(7, 5), sharex=True)

# Generate some sequential data
x = np.array(list("ABCDEFGHIJ"))
y1 = np.arange(1, 11)
sns.barplot(x=x, y=y1, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Sequential")

# Center the data to make it diverging
y2 = y1 - 5.5
sns.barplot(x=x, y=y2, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Diverging")

# Randomly reorder the data to make it qualitative
y3 = rs.choice(y1, len(y1), replace=False)
sns.barplot(x=x, y=y3, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Qualitative")

# Finalize the plot
sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=2)
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 
data = [23, 17, 35, 29, 12, 41]
 
 
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
 
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
 
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
 
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
 
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
 
# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
 
# show plot
plt.show()
```
